#TASK 2 : Loops (Multiples of a Number)
<H3>Code :</h3>


```python
def print_multiples_for(number):
    print(f"First 10 multiples of {number} using for loop:")
    for i in range(1, 11):
        print(number * i)

# Example execution
print_multiples_for(5)
```

<h3>Python Implementation Using while Loop :<h3>

```python
def print_multiples_while(number):
    print(f"First 10 multiples of {number} using while loop:")
    i = 1
    while i <= 10:
        print(number * i)
        i += 1

# Example execution
print_multiples_while(5)
```

<h3>Analysis of Loop Logic</h3>
<h3><b>For Loop:</b></h3>
The for loop uses the range(1, 11) function to iterate exactly ten times. This approach is concise and reduces the chances of errors because the number of iterations is predefined. It is generally preferred when the number of repetitions is known in advance.

<h3><b>While Loop:</b></h3>
The while loop executes as long as the condition i <= 10 remains true. A counter variable is initialized before the loop and incremented during each iteration. This structure provides more flexibility and is useful when the number of iterations depends on a condition rather than a fixed range.
<h3>Comparison of Looping Approaches :</h3>
<h3>Readability:</h3>
The for loop is more readable and compact, making it easier for beginners to understand. The while loop requires additional lines of code for initialization and incrementing the counter.

<h3>Control:</h3>
The while loop offers greater control over the looping condition and is suitable for scenarios where the number of iterations is uncertain.

<h3>Error Handling:</h3>
The for loop is less prone to logical errors such as infinite loops because it automatically stops after the specified range. In contrast, forgetting to increment the counter in a while loop can cause the program to run indefinitely.

<h3>Conclusion:</h3>
Both loops produce the same output; however, the for loop is typically recommended for fixed iterations due to its simplicity and reliability, while the while loop is better suited for condition-based execution.

# TASK 3 : Conditional Statements(Age Classisfication)
<h3>Code : </h3>

```python
def classify_age(age):
    if age < 13:
        return "Child"
    elif age >= 13 and age <= 19:
        return "Teenager"
    elif age >= 20 and age <= 59:
        return "Adult"
    else:
        return "Senior"


# Example execution
print(classify_age(10))
print(classify_age(16))
print(classify_age(35))
print(classify_age(65))
```

<h3>Alternative Implementation Using Dictionary-Based Logic :</h3>

```python
def classify_age_dict(age):
    age_groups = {
        range(0, 13): "Child",
        range(13, 20): "Teenager",
        range(20, 60): "Adult",
        range(60, 150): "Senior"
    }

    for age_range, group in age_groups.items():
        if age in age_range:
            return group


# Example execution
print(classify_age_dict(10))
print(classify_age_dict(16))
print(classify_age_dict(35))
print(classify_age_dict(65))
```

<h3>Analysis of Generated Conditions and Logic :</h3>
<h3>Simplified Logic Advantage:</h3>
The simplified version removes redundant comparisons by relying on the fact that earlier conditions have already been evaluated. For example, after confirming that the age is not less than 13, the program only needs to check if it is below 20 to identify a teenager. This improves readability and reduces complexity.

<h3>Dictionary-Based Logic:</h3>
The dictionary approach maps age ranges to their respective categories, creating a cleaner and more organized structure. It is particularly useful when managing multiple classifications, as updates can be made easily without modifying several conditional statements.

<h3>Conclusion:</h3>
All three implementations produce correct results. The simplified conditional structure is generally recommended due to its clarity and efficiency, while the dictionary-based approach is beneficial for scalability and better organization in larger applications.

#TASK 4 : For and While Loops (Sum of First n Numbers)
<h3>Code : </h3>

```python
def sum_to_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total


# Example execution
print(sum_to_n(5))
print(sum_to_n(10))
```

<h3>Alternative Implementation Using Mathematical Formula</h3>
<h3>Code :</h3>


```python
def sum_to_n_formula(n):
    return n * (n + 1) // 2


# Example execution
print(sum_to_n_formula(5))
print(sum_to_n_formula(10))
```

<h3>Explanation and Comparison of Approaches</h3>
<h3>For Loop Approach:</h3>
This method iterates through each number from 1 to n and adds it to a running total. It is easy to understand and ideal for learners who are beginning to work with loops.

<h3>Mathematical Formula Approach:</h3>
The formula n(n + 1) / 2 directly computes the sum without iteration. This makes the algorithm significantly faster with a time complexity of O(1), as the result is obtained through a single calculation. It is the most efficient approach, especially for large values of n.

<h3>Conclusion:</h3>
All three implementations produce the correct result. The loop-based methods are valuable for understanding fundamental programming concepts, whereas the mathematical formula is the preferred choice when efficiency and performance are important.

#TASK 5 : Classes (Bank Account Class)
<H3>Code :</h3>

```python
class BankAccount:
    # Constructor to initialize account holder name and starting balance
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    # Method to deposit money into the account
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited: {amount}")
        else:
            print("Invalid deposit amount.")

    # Method to withdraw money from the account
    def withdraw(self, amount):
        if amount <= self.balance and amount > 0:
            self.balance -= amount
            print(f"Withdrawn: {amount}")
        elif amount > self.balance:
            print("Insufficient balance.")
        else:
            print("Invalid withdrawal amount.")

    # Method to check current account balance
    def check_balance(self):
        print(f"Current Balance: {self.balance}")


# Example execution
account1 = BankAccount("Manohar", 1000)

account1.check_balance()
account1.deposit(500)
account1.withdraw(300)
account1.check_balance()
```

<h3>Explanation of the Code :</h3>
<h3>Class Structure:</h3>
The BankAccount class represents a simple model of a bank account. It contains attributes such as account_holder and balance, along with methods to perform common banking operations like depositing, withdrawing, and checking the balance.

<h3>Constructor (__init__):</h3>
The constructor initializes the account holderâ€™s name and sets the initial balance. A default balance of zero is provided so that an account can be created even if no initial deposit is made.

<h3>Deposit Method:</h3>
The deposit() method adds money to the account only if the entered amount is positive. This validation prevents incorrect transactions and ensures that the balance remains accurate.

<h3>Withdraw Method:</h3>
The withdraw() method checks two conditions before processing the transaction:

The withdrawal amount must be greater than zero.

The account must have sufficient funds.
If both conditions are satisfied, the specified amount is deducted from the balance; otherwise, an appropriate message is displayed.

<h3>Check Balance Method:</h3>
The check_balance() method simply prints the current balance, allowing the user to track account activity after each transaction.

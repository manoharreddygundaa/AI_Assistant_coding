#Task 2 :AI-Based Code Completion for Loop-Based String Processing
#Prompt :
Generate Python code to count vowels and consonants in a string
using a loop.

```python
text = input("Enter a string: ")
vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for char in text:
    if char.isalpha():
        if char in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Vowels:", vowel_count)
print("Consonants:", consonant_count)
```

#Explanation
In this task, AI-generated code processes a string using a loop to count vowels and consonants. The program accepts a string from the user and initializes counters for vowels and consonants. A for loop is used to iterate through each character in the string. The isalpha() function ensures that only alphabetic characters are counted, ignoring spaces and special symbols. If the character belongs to the vowel set, the vowel counter is increased; otherwise, the consonant counter is incremented. This task highlights effective use of loops and conditional logic for string processing.

Vowels : 4
Consonents : 8

Ignored spaces and numbers using isalpha().

#Library Management System
#Prompt Used:
Generate a Python program for a library management system using classes, loops, and conditional statements.

```python
class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print("Book added successfully")

    def display_books(self):
        if not self.books:
            print("No books available")
        else:
            for book in self.books:
                print(book)

lib = Library()

while True:
    print("\n1. Add Book\n2. Display Books\n3. Exit")
    choice = input("Enter choice: ")

    if choice == "1":
        name = input("Enter book name: ")
        lib.add_book(name)
    elif choice == "2":
        lib.display_books()
    elif choice == "3":
        break
    else:
        print("Invalid choice")
```

#Explanation (Library Management System)

This task demonstrates an AI-generated library management system using classes, loops, and conditionals. A Library class is created to store book details, and methods are defined to add and display books. A loop-driven menu allows continuous interaction with the user until they choose to exit. Conditional statements handle user choices such as adding or viewing books. This task shows how AI can generate structured, object-oriented programs, while also requiring human review to improve features and handle edge cases.

# Task 4 :AI-Assisted Code Completion for Class-Based Attendance System
#Prompt : <br>
Generate a Python class to mark and display student
attendance using loops.

```python
class Attendance:
    def __init__(self):
        self.students = {}

    def mark_attendance(self, name, status):
        self.students[name] = status

    def display_attendance(self):
        for name, status in self.students.items():
            print(name, ":", status)

att = Attendance()
att.mark_attendance("Rahul", "Present")
att.mark_attendance("Anjali", "Absent")
att.display_attendance()
```

#Task 5 : AI-Based Code Completion for Conditional Menu Navigation (ATM)
#Prompt Used
Generate a Python program using loops and conditionals to simulate an ATM menu.

```python
balance = 10000

while True:
    print("\nATM Menu")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

    choice = input("Choose option: ")

    if choice == "1":
        print("Balance:", balance)
    elif choice == "2":
        amount = int(input("Enter deposit amount: "))
        balance += amount
    elif choice == "3":
        amount = int(input("Enter withdrawal amount: "))
        if amount <= balance:
            balance -= amount
        else:
            print("Insufficient balance")
    elif choice == "4":
        break
    else:
        print("Invalid option")
```

Prevents overdraft
